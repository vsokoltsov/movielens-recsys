# 1. Extract, Transform, Load

```python
import os
from pathlib import Path
import pandas as pd
```

```python
ROOT = Path(os.path.abspath('')).resolve().parents[0]
DATA = os.path.join(ROOT, "data")
INTERIM_DATA = os.path.join(DATA, "interim")
RAW_DATA = os.path.join(DATA, "raw")
MOVIELENS_PATH = os.path.join(RAW_DATA, "ml-1m")
```

```python
users = pd.read_csv(
    os.path.join(MOVIELENS_PATH, "users.dat"),
    sep="::",
    engine="python",   # нужен для многосимвольного разделителя
    names=["user_id", "gender", "age", "occupation", "zip"],
    encoding="latin-1" # иногда нужно для MovieLens
)

ratings = pd.read_csv(
    os.path.join(MOVIELENS_PATH, "ratings.dat"),
    sep="::",
    engine="python",
    names=["user_id", "movie_id", "rating", "timestamp"]
)

movies = pd.read_csv(
    os.path.join(MOVIELENS_PATH, "movies.dat"),
    sep="::",
    engine="python",
    names=["movie_id", "title", "genres"],
    encoding="latin-1"
)
```
